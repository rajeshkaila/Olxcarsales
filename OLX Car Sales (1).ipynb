```python
print(np.round(olxdf.Kilometers_Driven.describe(),2))
print("skewness",olxdf.Kilometers_Driven.skew())
print("kurtosis",olxdf.Kilometers_Driven.kurt())
```

```python
print(np.round(olxdf.Mileage.describe(),2))
print("skewness",olxdf.Mileage.skew())
print("kurtosis",olxdf.Mileage.kurt())
```

```python
olxdf.Fuel_Type.value_counts().plot(kind="pie",autopct="%.2f%%")
```

```python
olxdf.Seats.value_counts().plot(kind="pie",autopct="%.2f%%")
```

```python
ax=pd.crosstab(olxdf.Fuel_Type,olxdf.Transmission).plot(kind="bar",
                                                    stacked=True)
for i in ax.containers:
    ax.bar_label(i,color="red")
```

```python
ax=pd.crosstab(olxdf.Seats,olxdf.Fuel_Type).plot(kind="bar",
                                                    stacked=True)
for i in ax.containers:
    ax.bar_label(i,color="red")
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))
fig,ax=plt.subplots(3,1)
sns.histplot(olxdf.Kilometers_Driven,ax=ax[0])
sns.boxplot(olxdf.Kilometers_Driven,ax=ax[1],orient="h")
sns.kdeplot(olxdf.Kilometers_Driven,ax=ax[2])
```

```python
plt.figure(figsize=(20,10))
fig,ax=plt.subplots(3,1)
sns.histplot(olxdf.Mileage,ax=ax[0])
sns.boxplot(olxdf.Mileage,ax=ax[1],orient="h")
sns.kdeplot(olxdf.Mileage,ax=ax[2])
```

```python
plt.figure(figsize=(20,10))
fig,ax=plt.subplots(3,1)
sns.histplot(olxdf.Price,ax=ax[0])
sns.boxplot(olxdf.Price,ax=ax[1],orient="h")
sns.kdeplot(olxdf.Price,ax=ax[2])
```
